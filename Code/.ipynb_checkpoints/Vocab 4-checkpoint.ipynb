```python
import re
import numpy as np
import pandas as pd
import csv

from collections import OrderedDict
from operator import itemgetter 
```

```python
def get_vocab(ifile):
    vocab = {}
    with open(ifile) as f:
        reader = csv.reader(f)
        for row in reader:
            text_in_row = ' '.join(row[1:])
            text_in_row = ''.join(re.findall("[a-zA-Z0-9 '-]*", text_in_row)).lower()
            words_in_row = text_in_row.split()
            #print(words_in_row, '\n')
            for word in words_in_row:
                try:
                    vocab[word] = vocab[word] + 1
                except:
                    vocab[word] = 1

    ord_vocab = OrderedDict(sorted(vocab.items(), key = itemgetter(1), reverse = True))
    
    return ord_vocab
```

```python
get_vocab("../Datasets/amazon_review_full_csv/train.csv")
```

```python
def write_vocab_file(input_file, vocab_file, num_words=10000):
    ord_vocab = get_vocab(input_file)
    with open(vocab_file, 'w') as textfile:
        for i, word in enumerate(ord_vocab):
            if i < num_words:
                textfile.write(word + "\t" + str(i+1) + "\t" + str(ord_vocab[word]) + "\n")
```

```python
# Amazon Review (Full) Dataset
write_vocab_file("../Datasets/amazon_review_full_csv/train.csv", "./Output/Vocabulary2/amazon_review_full.txt")
```

## Reformatting of <span style="color:brown">Train/Valid/Test Files</span> as Matrices of IDs of the Words Included

```python
def get_vocab_dictionary(vocabfile, nwords=5000):
    # Get a dictionary from the vocabulary file (which contains the top 10k most common words in the training set)
    # The top 10k words conform the keys of the dictionary, while their corresponding IDs determine the value of each of these keys
    vocab_dict = {}
    with open(vocabfile, "r") as f:
        for line in f[:nwords]:
            (word, idnum) = line.split()[0:2]
            vocab_dict[word] = idnum
    return vocab_dict


def get_file_of_ids(ifile, ofile, vocab_dict):
    # Get all the datapoints (i.e., reviews) from the input file
    with open(ifile, 'r') as textfile:
        reviews = textfile.readlines()
    # Write the output file, which consist in a sequence of IDs of the words contained in each review 
    # (only the ones that exist in the vocabulary)
    with open(ofile, "w") as idtext:
        for review in reviews:
           # Preprocessing: Removal of punctuation, and lower case
            preproc_review = ''.join(re.findall("[a-zA-Z ]*", review)).lower()
            # Get all the words (list)
            rwords = preproc_review.split()
            # For each of the words in the review:
            for rword in rwords:
                # Determine if the word appears in the vocabulary 
                try:
                    idtext.write(vocab_dict[rword]) # That is the ID of the word
                    idtext.write(" ")
                except:
                    pass
            # Finally, write the class separated by a tab 
            # (since the tab consists on 4 spaces, and we already have one writtenfrom the previous 'for' cyclewe only add other 3 spaces)
            idtext.write("   ")
            idtext.write(review[-2]) #The last character before the change in line '\n' is the class
            idtext.write("\n")
```

```python
# Call the previous functions to generate the train, valid, and test text files that 
# contain a sequence of ids corresponding to the words in each review for both datasets

# YELP Dataset
yelp_dict = get_vocab_dictionary("./Output/Dataset/yelp-vocab.txt")
get_file_of_ids("./hwk3_datasets/yelp-train.txt", "./Output/Dataset/yelp-train.txt", yelp_dict)

get_file_of_ids("./hwk3_datasets/yelp-valid.txt", "./Output/Dataset/yelp-valid.txt", yelp_dict)
get_file_of_ids("./hwk3_datasets/yelp-test.txt", "./Output/Dataset/yelp-test.txt", yelp_dict)

# IMDB Dataset
imdb_dict = get_vocab_dictionary("./Output/Dataset/IMDB-vocab.txt")
get_file_of_ids("./hwk3_datasets/IMDB-train.txt", "./Output/Dataset/IMDB-train.txt", imdb_dict)
get_file_of_ids("./hwk3_datasets/IMDB-valid.txt", "./Output/Dataset/IMDB-valid.txt", imdb_dict)
get_file_of_ids("./hwk3_datasets/IMDB-test.txt", "./Output/Dataset/IMDB-test.txt", imdb_dict)
```
